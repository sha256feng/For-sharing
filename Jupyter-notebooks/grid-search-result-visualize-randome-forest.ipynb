# Visualize the grid search result with Matplotlib

  Hyperparameter search using "GridSearchCV" from `sklearn` returns best model as well as the `cv_result_`. Here we use Matplotlib to visualize the grid search results and establish a rationale for hyperparameter tuning.

## Load the pickled result
The `cv_result_` was stored in binary format by `pickle` module. First load the data:

```python
import pickle
```

```python
load_result = pickle.load(open("grid_search/krr-cv-result-property-1.sav", 'rb'))
```

```python
# Print the loaded result
load_result
```

We use Pandas DataFrame to read the dictionary result of `cv_result` for better visualization of numbers and entries:

```python
import pandas as pd
from pandas import DataFrame as df
```

```python
df_data = df.from_dict(load_result)
```

```python
df_data
```

## Visualization in plot
Pick up 3 entries to do the plot:  
    1. alpha values;
    2. gamma values;
    3. corresponding test score;
Use 3D plotin Matplotlib to visualize the data;

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Read param alpha grid from dataframe
x = np.asarray(df_data['param_alpha'])
# x is an object, to enable log10() process, convert it to float
x = x.astype(np.float64)

# Read param gamma grid from dataframe
y = np.asarray(df_data['param_gamma'])
y = y.astype(np.float64)

# Read mean test score, neg MAE for each params grid
z = np.asarray(df_data['mean_test_score'])

# Do 3D scatter plot, x-y uses log scale
ax.scatter3D(np.log10(x), np.log10(y), z, c=z, cmap='copper')

# Labels
ax.set_xlabel('Alpha')
ax.set_ylabel('Gamma')
ax.set_zlabel('Mean test score: neg MAE')

plt.show()
```
